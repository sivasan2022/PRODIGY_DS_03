DATA VISUALIZATION

```python
# Top 10 countries with highest population in 2022
top_pop_2022 = df.nlargest(10,'2022 Population')
c=top_pop_2022['Country/Territory']
print(c)
percent = (top_pop_2022['2022 Population'] /
                  top_pop_2022['2022 Population'].sum()) * 100
print(percent)

plt.figure(figsize=(10, 6))
plt.bar(c, percent, color ='teal',
        width = 0.4)
plt.show()
```

```python
#Bar plot for top 10 countries with highest population in 1970
top_pop_1970 = df.nlargest(10,'1970 Population')
c1970=top_pop_1970['Country/Territory']
print(c1970)
percent1970 = (top_pop_1970['1970 Population'] /
                  top_pop_1970['1970 Population'].sum()) * 100
print(percent1970)

plt.figure(figsize=(10, 6))
plt.bar(c1970, percent1970, color ='pink',
        width = 0.4)
plt.show()
```

```python
print('Top 10 country by the population percentage:')
df.sort_values(by = 'World Population Percentage', ascending = False).head(10)
```

```python
# Looking top 10 country with largest area
print('Country with largest area:')
df.sort_values(by = 'Area (km²)', ascending = False).head(10)
```

```python
# Looking at top 10 countries with highest density
print('Country with highest density:')
df.sort_values(by = 'Density (per km²)', ascending = False).head(10)
```

```python
# Looking at top 10 countries with highest growth rate
print('Country with highest growth rate:')
df.sort_values(by = 'Growth Rate', ascending = False).head(10)
```

```python
# First we group the population by continent
print('We can already see Asia being the top of the list.')
df.groupby('Continent').sum().sort_values(by = 'World Population Percentage', ascending = False)
```

```python
# Now we reorder the columns and putting only the population in new dataframe
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', '2000 Population',
                               '2010 Population', '2015 Population', '2020 Population', '2022 Population',]].sum()
df2
```

```python
# Creating the line plot
plt.figure(figsize = (15, 6))
sns.lineplot(data = df2.transpose()) # Transpose to change the axis
plt.title("Population Growth by Continent (1970-2022)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(title = "Continent")
plt.show()
```

POPULATION ANALYSIS FOR INDIA

```python
# First we create a new dataframe sorts the country by decreasing of density
densityup = df.sort_values(by='Density (per km²)', ascending = False).reset_index()
densityup.index += 1
densityup.index.names = ['DenseRank']
densityup
```

POPULATION GROWTH OF INDIA

```python
#Creating line plot for India
pop = df.loc[92][['1970 Population', '1980 Population', '1990 Population', '2000 Population',
                     '2010 Population', '2015 Population', '2020 Population', '2022 Population',]]

plt.figure(figsize = (15, 6))
sns.lineplot(data = pop.transpose())
plt.title("Population Growth of India (1970-2022)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
